```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
pd.set_option('display.max_colwidth', None)
```

```python
page2 = requests.get("https://www.icc-cricket.com/rankings/womens/team-rankings/odi")
soup2 = BeautifulSoup(page2.content)
soup2
df=pd.DataFrame()
```

```python
def generic_team1(tag,class_tag,dfTag):
    women_team=soup2.find_all(tag,class_=class_tag)
    women_teams=[]
    for i in women_team:
        i=i.get_text().replace('\n'," ")
        i=i.strip(" ")
        women_teams.append(i)
    df[dfTag]=women_teams
generic_team1('span','u-hide-phablet','women_Team')
```

```python
df=df.drop(df.index[10:])
```

```python
def generic_team2(tag,class_tag,dfTag):
    women_team=soup2.find_all(tag,class_=class_tag)
    women_teams=[]
    for i in women_team:
        i=i.get_text().replace('\n'," ")
        i=i.strip(" ")
        women_teams.append(i)
    df[dfTag][1:10]=women_teams
```

```python
def generic_first1(tag,class_tag,dfTag):
    first_matches=soup2.find(tag,class_=class_tag)
    first_matches=first_matches.get_text()
    df[dfTag]=first_matches.replace("\n","")
generic_first1('td','rankings-block__banner--matches','women_team_matches')
generic_first1('td','rankings-block__banner--points','women_team_points')
generic_first1('td','rankings-block__banner--rating u-text-right','women_team_ratings')
```

```python
def split_values(tag,class_tag,dfTag1,dfTag2):
    women_teams_details=soup2.find_all(tag,class_=class_tag)
    women_teams_detail=[]
    for i in women_teams_details:
        i=i.get_text().replace('\n'," ")
        i=i.strip(" ")
        women_teams_detail.append(i)
    K = 2
    res = []
    for idx in range(0, K):
        res.append(women_teams_detail[idx::K])
    print("Stepped splitted List : " + str(res))
    result=list(res)
    df[dfTag1][1:10]=result[0]
    df[dfTag2][1:10]=result[1]
split_values('td','table-body__cell u-center-text','women_team_matches','women_team_points')
```
