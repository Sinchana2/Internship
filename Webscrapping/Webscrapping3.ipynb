#### 2. Scrape all the details from Amazon

```python
urls=[]
brand=[]
name=[]
rating=[]
count=[]
price=[]
delivery=[]
availability=[]
details1=[]
details2=[]
return1=[]
df_products=pd.DataFrame({})
```

```python
def product_search():

    for i in driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']"):
        urls.append(i.get_attribute("href"))

    for j in urls:
        driver.get(j)
        time.sleep(3)
        # Fetching Brand name
        try:
            brands = driver.find_element_by_xpath("//tr[@class='a-spacing-small']/td[@class='a-span3']/span[contains(text(),'Brand')]/parent::td/following-sibling::td/span")
            brand.append(brands.text)

        except:
            brand.append("-")
        # Fetching name of the product
        try:
            product_names = driver.find_element_by_xpath("//td[@class='a-span9']//span[contains(text(),'Ukulele')]")
            name.append(product_names.text)

        except:
            name.append("-")

        # Fetching ratings
        try:
            ratings = driver.find_element_by_xpath("//span[@class='a-size-medium a-color-base']")
            rating.append(ratings.text)

        except:
            rating.append("-")

        # Fetching Number of ratings
        try:
            counts = driver.find_element_by_xpath("//span[@id='acrCustomerReviewText']")
            count.append(counts.text)

        except:
            count.append("-")

        # Fetching Price
        try:
            prices = driver.find_element_by_xpath("//tr[@id='priceblock_ourprice_row']//td[2]//span[@class='a-size-medium a-color-price priceBlockBuyingPriceString']")
            price.append(prices.text)

        except:
            price.append("-")

        # Fetching expected delivery date
        try:
            deliveryD = driver.find_element_by_xpath("//div[@id='ddmDeliveryMessage']//b")

            delivery.append(deliveryD.text)

        except:
            delivery.append("-")

        # Fetching Availability
        try:
            availabilityD = driver.find_element_by_xpath("//span[@class='a-size-medium a-color-success']")

            availability.append(availabilityD.text)

        except:
            availability.append("-")

        # Fetching other details- color
        try:
            detailsD1 = driver.find_element_by_xpath("//tr[@class='a-spacing-small']/td[@class='a-span3']/span[contains(text(),'Colour')]/parent::td/following-sibling::td/span")

            details1.append(detailsD1.text)

        except:
            details1.append("-")

         # Fetching other details- material type   
        try:
            detailsD2 = driver.find_element_by_xpath("//tr[@class='a-spacing-small']/td[@class='a-span3']/span[contains(text(),'Back Material Type')]/parent::td/following-sibling::td/span")

            details2.append(detailsD2.text)

        except:
            details2.append("-")   

        try:
            returns = driver.find_element_by_xpath("//span[@class='a-declarative']//div[2]//a")

            return1.append(returns.text)

        except:
            return1.append("-")
for i in range(2,4):
    product_search()
    time.sleep(2)
    driver.get("https://www.amazon.in/s?k=ukelele&ref=nb_sb_noss_2")
    driver.find_element_by_xpath("//li[@class='a-normal']//a[contains(text(),'"+str(i)+"')]").click()
```

```python
df_products['Brand']=brand
df_products['Product Name']=name
df_products['Ratings']=rating
df_products['Number of ratings']=count
df_products['Price']=price 
df_products['Delivery day']=delivery
df_products['Availability']=availability
df_products['Return']=return1
df_products['Color']=details1
df_products['Material']=details2 
```

#### 8. Scrape billionaires details from forbes.in

```python
import selenium
import pandas as pd
import numpy as np
from selenium import webdriver
from selenium.webdriver.support.ui import Select
import time
```

```python
driver= webdriver.Chrome("C:/chromedriver.exe")
driver.get("https://www.forbes.com/billionaires/")
df=pd.DataFrame(columns=['Rank','Billionaires_Names','Networth','Age','Citizen','Source','Industry'])
```

```python
rank=[]
name=[]
networth=[]
age=[]
citizen=[]
source=[]
industry=[]
ranks=[]
```

```python
ranks = driver.find_elements_by_xpath("//div[@class='rank']")
for i in ranks:
    i = i.text
    i.replace(".",'')
    rank.append(i)
df['Rank']=rank

names = driver.find_elements_by_xpath("//div[@class='personName']//div")
for i in names:
    i = i.text
    name.append(i)
df['Billionaires_Names']=name


worth= driver.find_elements_by_xpath("//div[@class='netWorth']//div")
for i in worth:
    i=i.text
    i.replace(" ",'')
    networth.append(i)
K = 2
res = []
for idx in range(0, K):
    res.append(networth[idx::K])
result=list(res)
df['Networth']=result[0]


ages = driver.find_elements_by_xpath("//div[@class='age']//div")
for i in ages:
    i =i.text
    age.append(i)
df['Age']=age


citizenship = driver.find_elements_by_xpath("//div[@class='countryOfCitizenship']")
for i in citizenship:
    i=i.text
    citizen.append(i)
df['Citizen']=citizen


sources = driver.find_elements_by_xpath("//div[@class='source']//div//div")
for i in sources:
    i=i.text
    source.append(i)
K = 2
res = []
for idx in range(0, K):
    res.append(source[idx::K])
result=list(res)
df['Source']=result[0]


industries = driver.find_elements_by_xpath("//div[@class='category']//div")
for i in industries:
    i=i.text
    industry.append(i)
df['Industry']=industry
```
